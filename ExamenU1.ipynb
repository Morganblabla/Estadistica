```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/Morganblabla/Estadistica/refs/heads/main/mental_health.csv")
df.dropna(inplace = True)
df
```

```python
#a) Hipótesis de causalidad: Redacta una hipótesis sobre la causalidad entre las dos variables, a su vez,
#establece la variable dependiente y la variable independiente.

print("a) Hipótesis de causalidad: El aumento de los transtornos de ansiedad puede estar directamente relacionado con la prevalncia de transtornos depresivos.")
print("En pocas palabras, el aumento de los transtornos de ansiedad, puede llevar al incremento de transtornos depresivos.")

print("Variable dependiente: Transtornos depresivos")
print("Variable independiente: Transtornos de ansiedad")
x = df["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"]
y = df["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"]

#b) Recta de regresión ajustada: Realiza los cálculos necesarios para encontrar la recta de regresión ajustada. Incluye los procedimientos que te llevaron a la recta de regresión

from scipy.stats import linregress
#Se calcula la pendiente b1, que indica cuánto cambia y, por cada unidad de x.
#Se calcula la intersección b0, que es el valor de y cuando x=0.
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(f"Pendiente: {slope}, Intersección: {intercept}")
#Se suma la pendiente y la intersección y así se obtiene la recta de regresión.
print("La recta de regresión ajustada es: y = {:.2f}x + {:.2f}".format(slope, intercept))

#c) Grafica los puntos y la recta de regresión

import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x=df["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"], y=df["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"], scatter_kws={"color": "pink"}, line_kws={"color": "blue"})
plt.xlabel("Transtorno de Ansiedad")
plt.ylabel("Transtorno depresivo")
plt.title("Gráfico de dispersión con línea de regresión")
plt.grid()
plt.show()

#d) Coeficiente de correlación y determinación: Realiza los cálculos necesarios para obtener la r de Pearson y el coeficiente de determinación.
Sxx= np.sum((x- np.mean(x))**2)
Sxy= np.sum((x- np.mean(x))*(y- np.mean(y)))
Syy= np.sum((y- np.mean(y))**2)
r= Sxy/ np.sqrt(Sxx*Syy)
print("Coeficiente de correlacion:",r)
print("Coeficiente de determinacion",r**2)

#e) Interpreta los resultados del coeficiente de correlación y el coeficiente de determinación.
print("El coeficiente de correlación de 0.12038835725014921 indica que hay una relación muy débil entre las variables. Esto significa que el cambio en una variable no está fuertemente relacionado con el cambio en la otra.")
print("El coeficiente de determinación indica que solo el 1.45% de la variabilidad en la variable dependiente puede explicarse por la variable independiente.")
print("En general, estos resultados sugieren que no hay una relación lineal fuerte entre las variables.")

#f) Calcula el intervalo de confianza del 95% para β_1 y β_0 ¿Qué dice el intervalo de confianza a la posibilidad de que β_1 sea igual a cero?
b1 = Sxy / Sxx
b0 = np.mean(y) - b1 * np.mean(x)
Yc = b1 * x + b0
residuales = y - Yc
nivel_significancia = 0.05
from scipy.stats import t
t_critico = t.ppf(1 - nivel_significancia / 2, len(y) - 2)
se_b1 = np.sqrt(np.sum((y - Yc) ** 2) / (len(y) - 2)) / np.sqrt(Sxx)
confianza_b1 = (b1 - t_critico * se_b1, b1 + t_critico * se_b1)
print(f"Intervalo de confianza b1 al 95%: {confianza_b1}")
print("El intervalo no incluye 0, lo que sugiere que hay una relación significativa entre las dos variables.")
print("Básicamente sugiere que una variable si esta influyendo en la otra.")

#g) Realiza el gráfico de los residuales: A partir de este gráfico, menciona si los datos cumplen con los supuestos para la regresión:
#linealidad, normalidad en torno a la recta, homoscedasticidad

residuales= y-Yc
plt.figure()
plt.grid()
plt.scatter(x, residuales, color="pink")
plt.xlabel("Ansiedad")
plt.ylabel("Depresión")
plt.title("Grafico de residuales")
plt.axhline(y=0, color="blue", linestyle="--")

print("Linealidad: Los puntos están distribuidos aleatoriamente alrededor de la línea cero, lo que sugiere que la relación entre las variables es lineal.")
print("Normalidad en torno a la recta: Los puntos están distribuidos aleatoriamente alrededor de la línea cero, lo que sugiere que los residuos siguen una distribución normal.")
print("Homoscedasticidad: La dispersión de los puntos es constante a lo largo de la línea cero, lo que sugiere que la varianza de los residuos es constante.")
print("En resumen, los datos parecen cumplir con los supuestos para la regresión lineal.")
```
