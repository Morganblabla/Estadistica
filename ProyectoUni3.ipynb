```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')

# Entra a https://www.alphavantage.co/support/#api-key
# Registrate para obtener tu api-key

# Tu clave de API (reemplaza esta por la tuya)
my_key = 'OAD5N4A29WFBL7Z2.'   # esta es m√≠a, no la uses o habr√° problemas

# Crear objeto TimeSeries
ts = TimeSeries(key=my_key, output_format='pandas')

# Obtener datos diarios de AMZN
ticker = 'MSFT'
data, meta = ts.get_daily(symbol=ticker, outputsize='full')

columnas = [ticker + '_' + i for i in ['Open', 'High', 'Low', 'Close', 'Volume']]
data.columns = columnas

inicio = '2024-03-01'
fin = '2025-04-01'
df2 = data[(data.index >= inicio) & (data.index <= fin)]

df2
```

```python
# Comparativa
df = pd.merge(df1, df2, left_index=True, right_index=True)
df
```

2.- Se grafic√≥ los precios de cierre de Apple y Microsoft en el mismo gr√°fico, usando colores distintos para comparar visualmente su comportamiento a lo largo del tiempo.

```python
# Graficar
plt.figure(figsize=(12, 6))
plt.plot(df['AAPL_Close'], label='Apple', color='blue')
plt.plot(df['MSFT_Close'], label='MSFT', color='purple')
plt.title('Precios de Cierre de AAPL y MSFT')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre ($)')
plt.grid()
plt.legend()
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```

3.- Se aplicast√≥ la prueba de causalidad de Granger para ver si una acci√≥n predice a la otra.
Primero se analiza si Apple afecta a Microsoft. Luego se prob√≥ si Microsoft afecta a Apple.

```python
from statsmodels.tsa.stattools import grangercausalitytests

# Hip√≥tesis Nula (H‚ÇÄ): La serie X no causa en el sentido de Granger a la serie Y.

# Hip√≥tesis Alternativa (H‚ÇÅ): La serie X causa en el sentido de Granger a la serie Y.

# Definir el n√∫mero m√°ximo de rezagos para la prueba
max_lags = 7

# Realizar la prueba de causalidad de Granger
# La funci√≥n devuelve resultados para varios tests y cada rezago hasta el m√°ximo definido
resultado = grangercausalitytests(df[['AAPL_Close', 'MSFT_Close']], max_lags, verbose=True)
```

Se acepta la Ho hasta el d√¨a 6 y 7. Apple afecta a Microsoft.

```python
from prophet import Prophet

# Crear el modelo y ajustarlo
modelo = Prophet()
modelo.fit(df1['AAPL_Close'].reset_index().rename(columns={'date': 'ds', 'AAPL_Close': 'y'}))

# Predicci√≥n para los pr√≥ximos 365 d√≠as
futuro = modelo.make_future_dataframe(periods=365)
predicciones = modelo.predict(futuro)

# Visualizar las predicciones
fig = modelo.plot(predicciones)
plt.plot(df1['AAPL_Close'], label='Serie Original', color='red')
plt.xlabel('Fecha')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.title('Predicci√≥n con Prophet')
plt.ylabel('Precio de Cierre ($)')
plt.legend()
fig = modelo.plot_components(predicciones)
```
