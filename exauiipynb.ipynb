# f) Verifica los supuestos: linealidad, normalidad y homoscedasticidad

# Linealidad

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

X = df[['medida_cintura', 'peso', 'masa_corporal']]
Y = df['riesgo_hipertension']

X_const = sm.add_constant(X)

modelo_logistico = sm.Logit(Y, X_const)
resultado = modelo_logistico.fit()

predicciones = resultado.predict(X_const)

residuos_devianza = Y - predicciones

plt.figure(figsize=(8, 6))
plt.scatter(predicciones, residuos_devianza, color='pink')
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicciones (Probabilidades)")
plt.ylabel("Residuos de Devianza")
plt.title("Verificación de Linealidad en Regresión Logística")
plt.show()
```

El gráfico muestra que la relación log-odds entre las variables no está perfectamente lineal, pero no hay una desviación dramática que indique un modelo completamente incorrecto.

# Normalidad

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

residuos_devianza = Y - predicciones

plt.figure(figsize=(8, 6))
stats.probplot(residuos_devianza, dist="norm", plot=plt)
plt.title("Gráfico Q-Q de los Residuos de Devianza")
plt.show()
```

El gráfico Q-Q muestra que los residuos no se distribuyen normalmente. Las desviaciones en las colas del gráfico sugieren que los residuos pueden tener una distribución diferente a la normal, posiblemente con colas más gruesas (es decir, los valores extremos son más frecuentes de lo que se esperaría en una distribución normal).

# Homoscedasticidad

```python
import statsmodels.stats.api as sms

# Prueba de heteroscedasticidad de Breusch-Pagan
test_bp = sms.het_breuschpagan(residuos_devianza, X_const)
p_value_bp = test_bp[1]  # El valor p de la prueba

print(f"Valor p de Breusch-Pagan: {p_value_bp:.4f}")
```

 Hay evidencia significativa de heteroscedasticidad en los residuos de tu modelo. Esto significa que la varianza de los residuos no es constante a lo largo de las predicciones, lo que viola el supuesto de homoscedasticidad.
